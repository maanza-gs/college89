Generating the Indian Flag

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, RegularPolygon

width, height = 600, 400

saffron_color = [255 / 255, 153 / 255, 51 / 255]
white_color = [255 / 255, 255 / 255, 255 / 255]
green_color = [19 / 255, 136 / 255, 8 / 255]
navy_blue_color = [0 / 255, 56 / 255, 168 / 255]

flag = np.zeros((height, width, 3), dtype=np.float32)

stripe_height = height // 3
flag[:stripe_height, :] = saffron_color
flag[stripe_height:2 * stripe_height, :] = white_color
flag[2 * stripe_height:, :] = green_color

chakra_radius = stripe_height // 2
chakra_center = (width // 2, stripe_height // 2 + stripe_height)

num_spokes = 24
angle_increment = 360.0 / num_spokes
linewidth = 3

outer_circle = Circle(chakra_center, chakra_radius, edgecolor=navy_blue_color, facecolor='none', linewidth=linewidth)
plt.gca().add_patch(outer_circle)

for i in range(num_spokes):
    angle_rad = np.deg2rad(i * angle_increment)
    x = chakra_center[0] + chakra_radius * np.cos(angle_rad)
    y = chakra_center[1] + chakra_radius * np.sin(angle_rad)
    spoke = plt.Line2D([chakra_center[0], x], [chakra_center[1], y], color=navy_blue_color, linewidth=linewidth)
    plt.gca().add_line(spoke)

plt.imshow(flag)
plt.axis('off')
plt.show()
```

**Generating an elipse**

Important terms:
1. Major axis (a)
2. Minor axis (b)

Formula of an elipse:

$ \frac {(x-h)^2}{a^2} + \frac {(y-x)^2}{b^2} = r^2$

```python
img=np.zeros([256,256])
m,n=img.shape
r=5 #radius
a=5
b=10

for i in range(0,m):
  for j in range(0,n):
    if(i-128)**2/(a**2)+(j-128)**2/(b**2)<=r**2:
      img[i,j]=255

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
#Task2

# Create a blank image
image_size = 400
image = np.zeros((image_size, image_size, 3), dtype=np.uint8)

# Define center coordinates
center_x = image_size // 2
center_y = image_size // 2

# Number of concentric circles
num_circles = 5

# Radii of concentric circles
radii = np.linspace(20, 100, num_circles)  # Example: from radius 20 to 100, 5 circles

# Colors for concentric circles (optional)
colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (255, 255, 0), (255, 0, 255)]

# Draw concentric circles
for i in range(num_circles):
    radius = int(radii[i])
    color = colors[i % len(colors)]  # Cycle through colors
    theta = np.linspace(0, 2*np.pi, 100)  # Angle from 0 to 2*pi
    x = center_x + radius * np.cos(theta)
    y = center_y + radius * np.sin(theta)
    image[np.round(y).astype(int), np.round(x).astype(int)] = color

# Display the image using matplotlib
plt.imshow(image)
plt.title('Concentric Circles')
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```
