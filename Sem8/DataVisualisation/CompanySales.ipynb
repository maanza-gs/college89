```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Academics/Sem8/DataVisualisation/company_sales_data.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['month_number'], df['total_profit'], linestyle='--', color='red', marker='o', markersize=5, linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title('Total Profit of All Months with Style Properties')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for product in df.columns[1:7]:
    plt.plot(df['month_number'], df[product], label=product)
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title('Sales Data of All Products')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['month_number'], df['toothpaste'], color='blue')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales Data')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['month_number'], df['facecream'], label='Face Cream Sales')
plt.bar(df['month_number'], df['facewash'], bottom=df['facecream'], label='Face Wash Sales')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.title('Face Cream and Face Wash Sales Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['month_number'], df['bathingsoap'])
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.title('Bathing Soap Sales Data')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['total_profit'], bins=10)
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit')
plt.show()
```

```python
df_last_year = df[df['month_number'] == 12]
plt.figure(figsize=(10, 6))
plt.pie(df_last_year.iloc[:, 1:7].sum(), labels=df_last_year.columns[1:7], autopct='%1.1f%%')
plt.title('Total Sales Data for Last Year for Each Product')
plt.show()
```
